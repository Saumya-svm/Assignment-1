```julia
T = float(input("Kindly enter the temperature at which you want the properties"))
R = 8.314

def calculate_specific_heat(T):
    if T > species_properties[species]['common_temp']:
        coeff = species_coefficients[species]['low_temp']
    else:
        coeff  = species_coefficients[species]['high_temp']
    # cp = 0.0
    # for i in range(1, 7):
    #     cp += coeff['a'+str(i)] * (T**(i-1))
    # print(cp)
    # return cp
    return R*(coeff['a1'] + coeff['a2']*T + coeff['a3']* (T**2) + coeff['a4']*(T**3) + coeff['a5']*(T**4) + coeff['a6']*(T**5))

def calculate_enthalpy(T):
    if T > species_properties[species]['common_temp']:
        coeff = species_coefficients[species]['high_temp']
    else:
        coeff  = species_coefficients[species]['low_temp']
    return R*T*(coeff['a1'] + coeff['a2']*(T)/2 + coeff['a3']*(T**2)/3 + coeff['a4']*(T**3)/4 + coeff['a5']*(T**4)/5 + coeff['a6']/T)

def calculate_entropy(T):
    if T > species_properties[species]['common_temp']:
        coeff = species_coefficients[species]['high_temp']
    else:
        coeff  = species_coefficients[species]['low_temp']
    return R*(coeff['a1']*log(T) + coeff['a2']*T + coeff['a3']* (T**2)/2 + coeff['a4']*(T**3)/3 + coeff['a5']*(T**4)/4 + coeff['a7'])

for specie in species_properties:
    global species
    species = specie
    species_properties[specie]['specific_heat'] = calculate_specific_heat(T)    
    species_properties[specie]['enthalpy'] = calculate_enthalpy(T)    
    species_properties[specie]['entropy'] = calculate_entropy(T)    

```
